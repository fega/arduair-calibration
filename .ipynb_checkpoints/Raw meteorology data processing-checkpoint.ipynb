# ARDUAIR DATA COLLECTED PROCESSING

## Library imports

```python
import pandas as pd
import numpy as np
import datetime as dt
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## Data Processing

```python
#Lux bad values (-1) replaced with NaN
data=pd.read_csv('DATA.TXT',names=['year','month','day','hour','minute','second','Hum','Temp','Pr','L','pm10','pm25','o3','co','so2','no2'])
data['LNaN']=data['L'].replace(-1.00,np.nan).replace(0.00,np.nan)
data['L']=data['LNaN']
#Dates to datetime
dates=data[['year','month','day','hour','minute','second']]
dates['year']=dates['year'].add(2000)
dates['minute']=dates['minute'].add(60)
data['datetime']=pd.to_datetime(dates)

#agregation
data=data[['datetime','Hum','Temp','Pr','L','pm10','pm25','o3','co','so2','no2']]
data=data.groupby(pd.Grouper(key='datetime',freq='1h',axis=1)).mean()
# save
data.to_csv('ARDUAIR_ACROPOLIS_PROCESADO.csv')
# reset index
#data.reset_index()
#dt=data.groupby(['D','h']).mean()
data
```
