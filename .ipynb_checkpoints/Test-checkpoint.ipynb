```python
import pandas as pd
import numpy as np
import datetime as date
import seaborn as sns

def printme(h,t,p,l):
    print("""
H: {}
T: {}
P: {}
L: {}

    """.format(h,t,p,l))
    
```

## Importar los datos

```python
data = pd.read_csv('DATA.TXT',names=['Y','M','D','h','m','s','Hum','Temp','Pr','L','','','','','',''])
data.head()
```

## Dumb questions
### ¿Cuales son los promedios?:


```python
hum_prom=data['Hum'].mean()
temp_prom=data['Temp'].mean()
pr_prom=data['Pr'].mean()
l_prom=data['L'].mean()
printme(hum_prom,temp_prom,pr_prom,l_prom)

```

### ¿Valores maximos?

```python
max_hum = data[data['Hum'] == data['Hum'].max()]
max_temp = data[data['Temp'] == data['Temp'].max()]
max_pr = data[data['Pr'] == data['Pr'].max()]
max_l = data[data['L'] == data['L'].max()]
max_hum
```
