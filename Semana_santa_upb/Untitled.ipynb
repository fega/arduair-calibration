# Arduair data collected processing

## Library Imports

```python
import pandas as pd
import numpy as np
import datetime as dt
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

data=pd.read_csv('DATA.TXT',names=['year','month','day','hour','minute','second','Hum','Temp','Pr','L','co','so2','no2','pm10','pm25','o3'])
#filter wrong data
data=data[data.hour<24]

#Dates to datetime
dates=data[['year','month','day','hour','minute','second']]
dates['year']=dates['year'].add(2000)
dates['minute']=dates['minute'].add(60)
data['datetime']=pd.to_datetime(dates)

#agregation
data=data[['datetime','Hum','Temp','Pr','L','pm10','pm25','o3','co','so2','no2']]
data=data.groupby(pd.Grouper(key='datetime',freq='1h',axis=1)).mean()
plt.plot(data.pm10)
```

```python
plt.plot(data.pm25)
plt.plot(data.Temp*1000)
```
