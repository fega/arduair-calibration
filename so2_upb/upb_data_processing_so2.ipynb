# ARDUAIR: Procesamiento de datos SO2
Se procesan los datos para obtener un archivo CSV con los promedios hora de las lecturas de pm10  de las fechas 17, 18 y 19 de abril de 2017

```python
import pandas as pd
import numpy as np
import datetime as dt
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
pd.options.mode.chained_assignment = None 
```

```python
#load
df=pd.read_csv('DATA_raw.TXT',names=['year','month','day','hour','minute','second','hum','temp','pr','l','co','so2','no2','o3','pm10','pm25','void'])
df.head()

#Dates to datetime
dates=df[['year','month','day','hour','minute','second']]
dates['year']=dates['year'].add(2000)
dates['minute']=dates['minute'].add(60)
df['datetime']=pd.to_datetime(dates)

#discard non useful data
df_co=df[['datetime','co']]
df_so2=df[['datetime','so2']]
```

```python
sns.jointplot(data=df,x='so2',y='temp', kind='reg')
```

```python
sns.jointplot(data=df,x='co',y='temp', kind='reg')
```
